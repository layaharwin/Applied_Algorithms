```python
# Question 1:
import sys

def min_ones(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    
    # 1D array for storing ones
    DP=[]
    DP.append(0)
    DP.append(1)
    for i in range(2, n+1):
        DP.append(0)
    temp=sys.maxsize
    for i in range(2,n+1):
        DP[i]=temp
        for j in range(1,i):
            x= DP[j]+DP[i-j]
            if temp > x:
                temp=x
            if i%j==0:
                x= DP[j]+DP[i//j]
                if temp > x:
                    temp=x
            if str(i)[0:len(str(j))]==str(j):
                if str(i)[len(str(j))]!='0':
                    x= int(str(i)[len(str(j)):])
                    y= DP[j]+DP[x]
                    if temp > y:
                        temp=y
        DP[i]=temp
        temp=sys.maxsize
    return DP[n]

print(min_ones(22))
print(min_ones(9))
print(min_ones(1))
print(min_ones(0))
```

```python
# Question 3:
def get_max_coconuts(nums):
    if nums is None:
        return 
    
    # Creating 1D array for storing the max values
    DP=[]
    DP.append(0)
    DP[0]=nums[0]
    length=len(nums)
    for i in range(1, length):
        DP.append(0)
     
    for i in range(1,length):
        if i>=2:
            temp= DP[i-2]+nums[i]
            if temp > DP[i-1]:
                DP[i]= temp
            else:
                DP[i]= DP[i-1]
        else:
            if nums[i]>DP[i-1]:
                DP[i]= nums[i]
            else:
                DP[i]=DP[i-1]
    return DP[length-1]

print(get_max_coconuts([2,7,9,3,1]))#12
print(get_max_coconuts([50,1,8,100,200]))#258
print(get_max_coconuts([50,85,70,200,50]))#285
```

```python
# Question 5
def find_max_count(s,c1,c2):
    count_c1=0
    count_c2=0
    for i in range(0, len(s)):
        if s[i]==c1:
            count_c1=count_c1+1
        elif s[i]==c2:
            count_c2=count_c2+1      
    max_count=0
    if count_c1>count_c2:
        max_count=count_c1
    else:
        max_count=count_c2   
    return max_count
        
def get_max_sbstr(s,c1 ,c2):
    if c1 == c2 :
        max_count= find_max_count(s,c1,c2)
        return (max_count*(max_count+1))//2
    else:
        temp=0
        max_count= find_max_count(s,c1,c2)
        for i in s:
            if i==c1:
                temp+=1
            elif i==c2:
                max_count+=temp
        return max_count

print(get_max_sbstr("bcedecd",'b','d')) #4
print(get_max_sbstr("bcbe",'b','b')) #3
print(get_max_sbstr("aaaa",'a','a')) #10
```

```python
# Question 6
def num_passwords(digit_list):
    length= len(digit_list)
    if length ==0:
        return 0
    count={}
    count[digit_list[0]]=0
    # 1D array for tracking passwords
    DP=[]
    DP.append(1)
    DP.append(2)
    for i in range(2, length+1):
        DP.append(1)  
    for i in range(2,length+1):
        k=i-1
        DP[i]=2*DP[k];
        if digit_list[i-1] in count:
            DP[i]= DP[i]-DP[count[digit_list[k]]]
        count[digit_list[k]]=k
    out= DP[length]-1 
    return out

print(num_passwords([1,2,1,3])) #13
print(num_passwords([9, 9, 9, 9])) #4
print(num_passwords([])) #0
```
