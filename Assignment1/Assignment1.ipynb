```python
#Question 1:
def count_prime_letters(n):  
    if n <= 2:
        return "No Prime numbers found below this number"
    convert_num_words= {1:'one', 2:'two', 3:'three',4:'four',5:'five', 6: 'six', 7: 'seven', 8: 'eight', 9: 'nine', 10: 'ten', 11: 'eleven', 12: 'twelve', 13: 'thirteen', 14: 'fourteen', 15: 'fifteen', 16: 'sixteen', 17: 'seventeen', 18: 'eighteen', 19: 'nineteen', 20: 'twenty', 30: 'thirty',40: 'forty', 50: 'fifty', 60: 'sixty', 70: 'seventy', 80: 'eighty', 90: 'ninety'}
    count = 0
    #Iterate over all numbers from 2 to n
    for num in range(2,n):
        # Check prime
        temp=0
        for i in range(2, num//2 +1):
            if num % i == 0:
                #Not a prime number
                temp=1;
                break
        if temp==0 : 
            # prime
            # count= count + len( convert_num_words[num] if num<=19 else convert_num_words[(num/10)*10] + "" if num%10 ==0 else convert_num_words[num%10])
            if (num < 20) or (num % 10 == 0):
                count = count + len(convert_num_words[num])
            else:
                count = count + len(convert_num_words[(num//10)*10]) + len(convert_num_words[num%10])               
    return count
```

```python
# Question 2: 
def get_square_sum(M,N):
    output = 0;
    i=1;
    for j in range(1,M+1): # loop for M  j1
        output=0
        for k in range(1,N+1):   # range from 1 to t j
            while(i<=((k//2)+1)): # Checking for all values less than k/2 + 1
                if(i**2==k): # found perfect squares
                    output =output+k
                i+=1    
            i=1
        N=output
    return output
```

```python
# Question 3:
def check_bit(N):
    count=0;
    for i, value in enumerate(bin(N)[::-1]): # iterating through the binary string of N 
        if value=="1": # incrementing by 1, whenever 1 is encountered 
            count=count+1
        if count==2 :
            return i+1 # position of 2nd 1 found. 
    return -1
```

```python
# Question 4:
def find_multiple(n):
    # Here, the final result will 7 * 11 * 13 * n, However we are not allowed to use *
    # The first common multiple of 7, 11, 13 = 7 * 11 * 13 =  1001 which is fixed for all cases. 
    # According to 7,11 and 13 mystery, any number multiplied with 1001 will be same number written twice in the exact same order
    # eg 1 : 123 * 1001 = 123,123
    # eg 2 : 786 * 1001 = 786,786
    # since n is a 3 digit prime number in our case, the first common multiple will be 1001* n 
    return ((int)(str(n)+str(n)))  # Here n is converted to string twice and added together. Finally returned after converting into int. 
```

```python
#Question 5
def find_the_winner(matches_won , matches_drawn , goals_scored):

    points_list = [] # for couting points of each team
    no_of_teams = len(matches_won)
    for i in range(no_of_teams): # calculating points for each team
        points_list.append((matches_won[i]*3)+matches_drawn[i])
    
    # Unique Highest Score
    if points_list.count(max(points_list))== 1:
        return points_list.index(max(points_list))
    
    # Multiple Highest Score
    else:
        indices_w_same_points = [idx for idx, val in enumerate(points_list) if val == max(points_list)]
        temp_highest_goal = goals_scored[indices_w_same_points[0]]
        for i in indices_w_same_points:
            if temp_highest_goal < goals_scored[i]:
                temp_highest_goal = goals_scored[i]
        return goals_scored.index(temp_highest_goal)
```

```python
#Question 6 
def decoded_String ( S ) :
    dict_out={} # dictionary to count the characters
    decoded_string="" # output string 
    j=0;
    #decoded_String=""
    for i in S:
        if i in dict_out:
            dict_out[i] += 1 # increamenting count if found
        else :
            dict_out[i]=1  # adding to dictionary if not found 
        if dict_out[i]!= 3:
            decoded_string+=i # adding element to output string whne count is not 3
        else :
            dict_out[i]=0 # setting to 0, to take care of next 3rd occurance of character i
        j=j+1
    return ( decoded_string )
```
