```python
# Problem 1:

# Function for merging the right and left side
def merge_left_right(clock, left, mid, right):
    i = 0     
    j = 0   
    temp = left 
    temp1 = [0] * (mid - left + 1)
    temp2 = [0] * (right - mid)
    for index in range(0, mid - left + 1):
        temp1[index] = clock[left + index]
    for index in range(0, right - mid):
        temp2[index] = clock[mid + 1 + index]   
    while i < mid - left + 1 and j < right - mid:
        if temp1[i] <= temp2[j]:
            clock[temp] = temp1[i]
            i += 1
        else:
            clock[temp] = temp2[j]
            j += 1
        temp += 1
    while i < mid - left + 1:
        clock[temp]= temp1[i]
        i += 1
        temp += 1
    while j < right - mid:
        clock[temp] = temp2[j]
        j += 1
        temp += 1

# Function for merge sort
def merge_Sort(clock, left, right):
    if left < right:
        mid = left+(right-l)//2
        merge_sort(clock, left, mid)
        merge_sort(clock, mid+1, right)
        merge_left_right(clock, left, mid, right)
        return clock
    
def maxEmployees(clock_in, clock_out):
    length= len(clock_in)
    if (length==1):
        return (1,1)
    # Merge sort on clock_in -> Complexity is O(nlogn)
    clock_in_1= mergeSort(clock_in, 0, length-1)
    # Merge sort on clock_out -> Complexity is O(nlogn)
    clock_out_2= mergeSort(clock_out, 0, length-1)
    clock_in_out_1 = [[x, 1] for x in clock_in_1] + [[x, -1] for x in clock_out_2]
    i=0
    j=length
    clock_in_out_2=[]
    # Merging the lists-> Complexity is O(n)
    while(i <length and j <2*length): 
        if clock_in_out_1[i][0]< clock_in_out_1[j][0]:
            clock_in_out_2.append(clock_in_out_1[i])
            i=i+1
        elif clock_in_out_1[i][0]> clock_in_out_1[j][0]:
            clock_in_out_2.append(clock_in_out_1[j])
            j=j+1
        elif (clock_in_out_1[i][0]== clock_in_out_1[j][0]):
            if clock_in_out_1[i][1]> clock_in_out_1[j][1]:
                clock_in_out_2.append(clock_in_out_1[i])
                clock_in_out_2.append(clock_in_out_1[j])
            else:
                clock_in_out_2.append(clock_in_out_1[j])
                clock_in_out_2.append(clock_in_out_1[i])
            i=i+1
            j=j+1
    clock_in_out= clock_in_out_2
    out = []
    total = 0
    temp = 0
    # Calculating the time interval with max employees-> Complexity is O(n)
    for i, j in clock_in_out:
        temp += j
        if temp > total:
            total = temp
            out = []
            out.append(i)
        elif temp == total:
            out.append(i)
    return (total, len(out))

print(maxEmployees([1, 2, 4, 7, 8, 12], [3, 7, 8, 12, 10, 15 ])) #(3,2)
print(maxEmployees([2,2,3,5,6,7,10], [4,5,7,10,6,10,11])) #(3,5)
print(maxEmployees([1,2], [1,4])) #(1,2)
```

```python
# Problem 2:

def find_largest_integer(num):
    out= [int(x) for x in str(num)]
    index=-1
    # Checking in reverse sord if its sorted, if not store where its breaking and stroing the index in index variable 
    # Complexity of loop is O(n)
    for i in range(len(out)-1,-1, -1):
        if out[i] < out[i-1]:
            index= i-1
            value=out[index]
            break
    # If index is changed, then there exists a max value less than num
    if index != -1 : 
        new_index=-1
        value=-1
        # Finding max value less than num[index]
        # Complexity of loop is O(n)
        for i in range(len(out)-1, index, -1):
            if out[i] > value and out[i]< out[index]:
                new_index=i
                value=out[i]
        # Swapping the index and new_index for final output
        out[index], out[new_index] = out[new_index], out[index]
    # Converting list to int
    s = [str(i) for i in out] 
    res = int("".join(s))
    return res

print(find_largest_integer(12435))#12345
print(find_largest_integer(18256)) #16258
print(find_largest_integer(572)) #527
print(find_largest_integer(21)) #12
print(find_largest_integer(12345)) #12345
```

```python
# Problem 3
def recur_quicksort(num, left, right):
    if left < right:
        temp=num[right]
        temp2=left-1
        for j in range(left, right):
            if num[j] <= temp:
                temp2 = temp2 + 1
                (num[temp2], num[j]) = (num[j], num[temp2])
        (num[temp2 + 1], num[right]) = (num[right], num[temp2 + 1])
        # sorting left side
        recur_quicksort(num, left, temp2+1 - 1)
        # sorting right side 
        recur_quicksort(num, temp2+2, right)
        return num

def quicksort(num):
    if len(num)==1:
        return num
    return recur_quicksort(num, 0, len(num)-1)

print(quicksort([7,7,3,2,1])) #[1,2,3,7,7]
print(quicksort([4,3,1,2])) #[1,2,3,4]
print(quicksort([10])) #[10]
print(quicksort([10,1])) #[1,10]
```

```python
# Problem 4 :
class Wavelet_Tree :
    
    def __init__( self , A) :
        # List for Wavelet Tree
        self.main_list=[]
        pos=['X','0','1', '2']
        list1=[]
        list1.append(A)
        
        while list1:
            temp=list1.pop(0)
            if temp[0]==[pos[2]]:
                if temp[1]!=[pos[0]]:
                    self.append_function(temp[1], pos[0])
                elif temp[1]==[pos[0]]:
                    self.append_function(pos[0], temp[2])
            
            elif temp==[pos[3]]: 
                self.append_function(pos[0],pos[0])
                continue 
            
            else:
                
                middle=(min(temp)+max(temp))/2 
                out=""
                
                for n in temp:
                    if n >middle:
                        out+=pos[2] 
                    else:
                        out+=pos[1] 
                
                left_child=[]
                for i in temp:
                    if i <=middle:
                        left_child.append(i)
                
                
                right_child=[]
                for i in temp:
                    if i > middle:
                        right_child.append(i)
                
                len_left= len(left_child)
                len_right=len(right_child)
                if len_left<=1 :
                    if len_right<=1:
                        self.main_list.append(out)
                        list1.append([pos[3]])
                        continue
                    
                elif len_left<=1 or len_right<=1:
                    self.main_list.append(out)
                    
                    if len_left<=1:
                        self.helper(right_child, list1, pos )
                        continue
                        
                    else:
                        self.helper2(left_child, list1, pos )
                        continue
                        
                self.main_list.append(out)
                list1.append(left_child)
                list1.append(right_child)
                
                
    def append_function(self, a, b):
        self.main_list.append(a)
        self.main_list.append(b)
    
    
    def helper(self, right_child, list1, pos ): 
        out=[]
        out.append([pos[2]])
        out.append([pos[0]])
        mid=(min(right_child)+max(right_child))/2 
        main_out=""
        for n in right_child:
            if n > mid:
                main_out+=pos[2]
            elif n <= mid:
                main_out+=pos[1]
        out.append(main_out)
        list1.append(out)  
                      
            
    def helper2(self, left_child, list1, pos ):
        out=[]
        out.append([pos[2]])
        middle=(min(left_child)+max(left_child))/2 
        main_out=""
        for n in left_child:
            if n >middle:
                main_out+='1'
            else:
                main_out+='0'
        out.append(main_out)
        out.append([pos[0]])
        list1.append(out)  
        
        
    def helper3(self,i,h,j,l,r):
        pos=['0','1']
        print(f'Level {h} :',(j,l,r))
        tmp= 2*i
        if l>=r:
            return
        if self.main_list[i][l-1:r].count(pos[0]) < j :  
            self.helper3(tmp+2,h+1, j-(self.main_list[i][l-1:r].count(pos[0])),self.main_list[i][:l-1].count(pos[1])+1 ,self.main_list[i][:r].count('1'))
        else:   
            self.helper3(tmp+1,h+1,j,self.main_list[i][:l-1].count(pos[0])+1 ,self.main_list[i][:r].count(pos[0]))
     
    
    def RQQ(self, k, left, right) :
        self.helper3(0,0,k,left,right)
        return
    
    
    def print(self) :
        i=0
        k=0
        while(i<len(self.main_list)):
            length= min(len(self.main_list),2**k+i)
            print(f'Level {k} : ',end = '')
            for j in range(i,length):
                if j != length-1:
                    print(self.main_list[j],', ',end="")
                elif j == length-1:
                    print(self.main_list[j])
            k+=1
            i=2**(k)-1
        return
    
wv_tree = Wavelet_Tree([6, 2, 0, 7, 9, 3, 1, 8, 5, 4])
wv_tree.print() 
wv_tree.RQQ(5, 3, 9)
```

```python
# Problem 5
import itertools   
def shortest_superstring(A):
    dict1={}
    out = ["".join(A)]
    
    for d1, d2 in itertools.combinations(A, 2):
        for d1, d2 in ((d1, d2), (d2, d1)):
            m=False
                
            for l in range(len(d2), 0, -1):
                if d1.endswith(d2[:l]):
                    m=True
                    break
                else:
                    continue
                    
            if m==False:
                l= 0  
            if d1 not in dict1 or l > dict1[d1][0]:
                dict1[d1] = [l, d2]
            elif l == dict1[d1][0]:
                dict1[d1].append(d2)
            else:
                continue
                    
    def helper(i, j, k):
        if len(i + "".join(k)) < len(out[0]):
            out[0] = i + "".join(k)
        if k and any(q in k for q in dict1[j][1:]):
            for q in dict1[j][1:]:
                if q in k:
                    helper(i + q[dict1[j][0]:], q, k - {q})
                    
        else:
            for q1 in k:
                for q in dict1[q1][1:]:
                    if q in k:
                        helper(i + q1 + q[dict1[q1][0]:], q, k - {q1, q})
                            
    for o in A:
        helper(o, o, set(A) - {o})
    return out[0]

print(shortest_superstring(["CATGC", "CTAAGT", "GCTA", "TTCA", "ATGCATC"])) #GCTAAGTTCATGCATC
print(shortest_superstring(["abcd", "bcde", "cdbc"])) #"abcdecdbc" when in fact the correct solution is "abcdbcde"
```
