```python
import math
class amor_dict ():
    def __init__(self, num_list):
        self.d={}
        l1= len(num_list)
        # Finding key value for num_list
        key = int(math.log(l1,2)) 
        #CASE 1 :
        # Number of elements is equal to 2**key
        # i.e. we dont have to split the elements into different levels and directly put them into one level
        if 2**key == l1:
            num_list.sort()
            # Creating empty list for all other levels other than key
            for i in range(0, key):
                self.d[i]=[]
            # Inserting num_list for key
            self.d[key]= num_list
        #CASE 2:
        # Number of elemenst is not equal to 2**key
        # Now we need to split the elements into multiple levels
        else :
            # Finding levels for inserting num_list
            levels = []
            while (l1 > 0):
                levels.append(int(l1 % 2))
                l1 = int(l1 / 2)
            # Inserting num_list elements into levels
            for i in range(0, len(levels)):
                # Inserting elements for key value i 
                if (levels[i] == 1):
                    self.d[i]=[]
                    for j in range(0, 2**i):
                        # Popping elements from num_list everytime its added into levels
                        self.d[i].append(num_list.pop())
                    self.d[i].sort()
            # Creating empty list for all other levels in which elements are not added
            for i in range(0, max(self.d.keys())):
                if i not in self.d:
                    self.d[i]=[]
                    
    def insert(self , num):
        #Case 1:
        #level 0 is emply and element is directly added
        if self.d[0]==[]:
            self.d[0].append(num);
        #Case 2:
        #Scenario 1: element added without creating a new level
        #Scenario 2: element added by creating a new level
        else: 
            #elementcount=0;
            positionfound=-1
            #Finding level where num needs to be added
            for i in range (0, max(self.d.keys())):
                if len(self.d[i])==0:
                    positionfound = i
                    break
                #else:
                    #elementcount+= len(self.d[i])
            #Case 2: Scenario 1: element added without creating a new level
            if positionfound>0: 
                self.d[positionfound].append(num);
                for i in range (0, positionfound):
                    #Adding all element in levels above positionfound to level-> positionfound
                    length= len(self.d[i])
                    for j in range (0, length):
                        self.d[positionfound].append(self.d[j].pop())
                self.d[positionfound].sort()
            #Case 2: Scenario 2: element added by creating a new level
            else:
                # New level needs to added to insert the element
                positionfound= max(self.d.keys())+1;
                self.d[positionfound]=[]
                self.d[positionfound].append(num);
                # Adding all elements above level positionfound to level-> positionfound
                for i in range (0, positionfound):
                    length= len(self.d[i])
                    for j in range (0, length):
                        self.d[positionfound].append(self.d[i].pop())
                self.d[positionfound].sort()
    
    def search(self , num):
        for i in range (0,max(self.d.keys())+1) :
            for j in range(0, len(self.d[i])):
                if self.d[i][j]== num:
                    return i;
        return -1
        
    def print(self):
        for i in range(0, max(self.d.keys())+1) :
            if self.d[i]==[]:
                print(i,":empty", sep="")
            else:
                print(i,":",self.d[i],sep="")


print("Case 1: Splitting the list into multiple levels")       
ad_obj = amor_dict([23, 12 ,24])
ad_obj.print()
print("")
print("Inserting 11")
ad_obj.insert(11)
ad_obj.print()
print("")
print("Inserting 13")
ad_obj.insert(13)
ad_obj.print()
print("")
print("Inserting 50")
ad_obj.insert(50)
ad_obj.print()
print("")
print("Inserting 51")
ad_obj.insert(51)
ad_obj.print()
print("")
print("Searching 13")
print(ad_obj.search(13))
print("")
print("Searching 77")
print(ad_obj.search(77))

print("\nCase 2: Adding list to single level")       
ad_obj = amor_dict([23, 12 ,24, 100])
ad_obj.print()
print("")
print("Inserting 11")
ad_obj.insert(11)
ad_obj.print()
print("")
print("Inserting 13")
ad_obj.insert(13)
ad_obj.print()
print("")
print("Inserting 50")
ad_obj.insert(50)
ad_obj.print()
print("")
print("Inserting 51")
ad_obj.insert(51)
ad_obj.print()
print("")
print("Searching 13")
print(ad_obj.search(13))
print("")
print("Searching 77")
print(ad_obj.search(77))
```

```python
def  count_digit_string(n):
    # n is 1, we return "1"
    if n==1:
        return "1"
    # Calling count_digit_string for n-1 always
    newstring = count_digit_string(n-1)
    count = 1
    output = ""
    # if the new string is 1 return count of 1 and 1 
    if newstring == "1":
        return "11"
    # else if new string is 11 return count of 1 and 1
    elif newstring == "11":
        return "21"
    i = 1 
    while (i < len(newstring)):
        length= len(newstring)-1
        # Incrementing the count when when numbers are same
        if newstring[i] == newstring[i-1]:
            count = count + 1
        # Adding count and string into output when newstring[i] != newstring[i-1] and initializing the count variable
        elif newstring[i] != newstring[i-1]:
            output = output + str(count)+newstring[i-1]
            count = 1
        # Adding count and string into output when length of newstring is i 
        if i == length:
            output = output + str(count)+newstring[i]
        i = i + 1
    return output

print("1: ",count_digit_string(1))
print("2: ",count_digit_string(2))
print("3: ",count_digit_string(3))
print("4: ",count_digit_string(4))
print("5: ",count_digit_string(5))
print("6: ",count_digit_string(6))
print("10:",count_digit_string(10))
print("30:",count_digit_string(30))
```
