## Problem 1
Fräulein Anja and Julia bought a rectangular Black Forest cake.
Black Forest cakes have cherries on top of them which they both love. Anja would like your help to cut the cake into two smaller rectangular pieces in such a way that she gets the maximum possible cherries after Julia takes the best piece (the one with more cherries) out of the two pieces.
Input will be a matrix representing the cake. Here ‘#‘ represents a cherry and ‘.‘ represents no-cherry (just cake). Your output will be the maximum possible number of cherries on Anja’s piece after Julia takes the piece with more cherries. Note that you are allowed to cut the cake only once, and the cut has to be either horizontal or vertical.

```python
def cake(input):
    no_of_rows = len(input)
    no_of_cols =  len(input[0])
    total_cherries = 0
    
    # Finding total cherries in the cake
    for i in range(no_of_rows):
        for j in range(no_of_cols):
            if input[i][j] == "#":
                total_cherries += 1
    
    #print("total_cherries:", total_cherries)
    
    if total_cherries % 2 == 0:
        min_cherry = total_cherries // 2
    else:
        min_cherry = (total_cherries // 2) + 1
    #print("min_cherry:", min_cherry)
    
    # Checking row wise: top down
    julia_row_cherry = 0
    for i in range(no_of_rows):
        for j in range(no_of_cols):
            if input[i][j] == "#":
                julia_row_cherry += 1
        if julia_row_cherry >= min_cherry:
                break
                
    anja_row_cherry = total_cherries - julia_row_cherry
    # Checking row wise: bottom up
    julia_row_cherry_2 = 0
    for i in range(no_of_rows-1, -1, -1):
        for j in range(no_of_cols-1, -1, -1):
            if input[i][j] == "#":
                julia_row_cherry_2 += 1
        if julia_row_cherry_2 >= min_cherry:
                break
    anja_row_cherry_2 = total_cherries - julia_row_cherry_2
    
    # Checking col wise: top down
    julia_col_cherry = 0
    for j in range(no_of_cols):
        for i in range(no_of_rows):
            if input[i][j] == "#":
                julia_col_cherry += 1
        if julia_col_cherry >= min_cherry:
                break
    anja_col_cherry = total_cherries - julia_col_cherry
    
    # Checking col wise: bottom up
    julia_col_cherry_2 = 0
    for j in range(no_of_cols-1, -1, -1):
        for i in range(no_of_rows-1, -1, -1):
            if input[i][j] == "#":
                julia_col_cherry_2 += 1
        if julia_col_cherry_2 >= min_cherry:
                break
    anja_col_cherry_2 = total_cherries - julia_col_cherry_2
    # finding max out of all
    max_cheery_on_anja_piece = max(anja_row_cherry, anja_row_cherry_2, anja_col_cherry, anja_col_cherry_2)
    
    return max_cheery_on_anja_piece 
```

A spy posted on a building can spy on people in the buildings to the right which are shorter than the building she is posted on (Of course, she can also spy on the people in her building).
You may assume that all buildings have distinct heights. Your input will be the heights of the buildings, and your output has to be the locations (indices) of the buildings on which to post spies.


```python
#Question 2:
def spy_locations(heights):
    if not heights:
        return []
    # Storing the index of first building as we always need a spy there
    spy =[0]
    # Stack is used to store the highest height 
    stack= (heights[0])
    for i, value in enumerate(heights[1:]):
        if stack< value : # When ever a new higher value is found, the value of stack is updated with it 
            stack=(value)
            spy.append(i+1)
    return spy;
```

## Problem 3
Veidt and Walter are taking turns while playing a game. Given a string pattern p of length n and a string s of length N>n on each turn a player has to remove a character either from the beginning or the end of the string s. Game stops when the size of string s becomes n. Walter wins if the pattern p and the resultant string s are identical. Else Veidt wins. Assuming Veidt always makes the first move and both the players always play using optimal strategy, return the winner of the game. Expected Time Complexity O(n)

```python
def whos_the_winner(s,p):
    s_length= len(s)
    p_length = len(p)
    for i in range(s_length):
        # Looking for the p's pattern in s and storing its index in i 
        if s[i:i+p_length] == p:
            index = i
    left = index
    right = (s_length - index) - p_length
    # when right index = left index , then walter wins as there will be eqaul distance from both the side of string and Walter will alwasy try to maintain it after Veidt's turn
    if left == right:
        winner = 'Walter'
    # else Veidt wins as there will be uneven distance on both sides of the string. Hence it will be easier for Veidt to reach p and cut the pattern
    else:
        winner = 'Veidt'
    return winner
```

```python
print(whos_the_winner("befg","ef"))
print(whos_the_winner("Abcdefghij", "cdef"))
print(whos_the_winner("asdfghjk","dfgh"))
print(whos_the_winner("asdfghjk","sdfg"))
print(whos_the_winner("bacab", "ac"))
print(whos_the_winner("baacb", "ac"))
print(whos_the_winner("baaab", "ab"))
print(whos_the_winner("abcdabacdb", "ab"))
```

## Problem 4
Given a pattern p and a text string s, remove the pattern from the string s whenever it’s instance is found as a substring and concatenate the remaining two parts of the string together. Repeat the above process till there are no more substrings in s that are equal to p and return the resultant string

```python
def return_resultant_string(s, p):
    s_length= len(s)
    p_length = len(p)
    # iterating till s string is present in s
    while p in s:
        for i in range(len(s)): 
            # finding the pattern and removing it 
            if s[i:i+p_length] == p:
                index = i
                s = s[0:index] + s[index+p_length:s_length]
    result = s
    return result
```

```python
print(return_resultant_string("aababccbc", "abc"))
print(return_resultant_string("aababccbcab", "abc"))
print(return_resultant_string("bacaababccbc", "abc"))
print(return_resultant_string("aababccbc", "abc"))
print(return_resultant_string("aababccbcab", "abc"))
print(return_resultant_string("bacaababccbc", "abc"))
```

## Problem 5
Given a binary string of n characters and a list of strings queries where each element in the list is either “flip” or “get”. For each element query in “queries”, if the query is “flip” you have to find the first “1” from the left, set it to ‘0’ and set all the ‘0’s before it to ‘1’.
Whenever the query is “get” you have to append the number of ‘1’s in the binary string to a “result” list. After accounting for all the queries return the list of integers result. The operations “flip” and “get” are to be implemented in O(1) complexity.
One may safely assume there is at least one ‘1’ in the string when either of the above functions is called.

```python
def flip_game(s,queries):
    # Creating lookup table
    dict_lookup= {}
    # Reversing the string and converting to binary
    dec = int(s[::-1],2)
    for i in range(0,len(queries)):
        if (dec<0):
            dec=0
        dict_lookup[i] = bin(dec).count("1")
        # Flip operation is equivalent to reverse of binary - 1
        dec = dec - 1
        
    #print(dict_lookup)
    result_list=[]
    # Using the lookup we can now do flip and get operation in O(1)
    k=0
    for i in queries:
        if i == "get":
            result_list.append(dict_lookup[k] )
        else:
            k=k+1
    return result_list
```

```python
print(flip_game("0000101011", ["get","flip","flip","get","flip","flip","flip","get"]))
print(flip_game("0010101011", ["get","flip","flip","get"]))
print(flip_game("000000000", ["get","flip","flip","get"]))
print(flip_game("100000",["flip","get","flip","get"]))
print(flip_game("0000111",["flip","get","flip","get"]))
```

## Problem 6
You are provided with a scoresheet, Calculate the total number of points scored at the end of the game by following the rules.
Rules:
1. An integer x - Record a new score of x.
2. [+, -, *, %, /]- Record a new score by performing a given operation on previous two scores. It is guaranteed there will always be two previous scores.
3. "D" - Record a new score that is double the previous score. It is guaranteed there will always be a previous score.
4. "I" - Invalidate the previous score, remove it from the record. It is guaranteed that there will always be a previous score.
Note: - For “/” operation consider the floor division of the given 2 numbers.

```python
def calculate_points (list):
    scoresheet=[]
    for i in list:
        if i== "I":
            scoresheet.pop()
        elif i == "D":
            scoresheet.append(int(scoresheet[-1])*2)
        elif i == "+": 
            scoresheet.append(int(scoresheet[-1]) + int(scoresheet[-2]))
        elif i == "-":
            scoresheet.append(int(scoresheet[-2] - int(scoresheet[-1])))
        elif i == "*":
            scoresheet.append(int(scoresheet[-1]) * int(scoresheet[-2]))
        elif i == "/":
            scoresheet.append(int(scoresheet[-2] / int(scoresheet[-1])))
        elif i == "%" :
            scoresheet.append(int(scoresheet[-1]) % int(scoresheet[-2]))
        else :
            scoresheet.append(int(i))
    #print(scoresheet)
    return sum(scoresheet)
```

```python
print(calculate_points(["5","2","I","D","+"]))
print(calculate_points(["5","-2","4","I","D","9","+","/"]))
print(calculate_points(["10","10","/","5","5","-","+"]))
print(calculate_points(["11","I","2","D","3","-"]))
print(calculate_points(["11","2","2","D","3","*","%"]))
```

# Problem 8
There is a colony of ants where each ant has been assigned a positive number as its identity. Ants returning from food hunt form a line outside the nest. Their nest has a single narrow entrance where only one ant, at a time can enter the nest through it.
The colony has a rule, where any ant with number ‘x’, arriving at the entrance cannot enter the nest before all the preceding numbered ants have entered (x-1,x-2,...) the nest [ant 4 cannot enter till ants 3,2 and 1 have entered the nest].
There is a narrow tunnel near the entrance where any number of ants can ‘wait’ if required, till it’s their turn to enter the nest. Any ant that enters the waiting tunnel last must necessarily leave the tunnel first, and any ant leaving the waiting tunnel can only go to the entrance of the nest, and cannot re-enter the arrival queue.

```python
def validate_order(arrival_queue):
    wait_tunnel = []
    nest = []
    counter = 1

  # Iterate over arrival queue
    for ant_i in arrival_queue:
        if ant_i == counter:
            counter += 1
            nest.append(ant_i)
            # Check wait_tunnel now
            while wait_tunnel:
                last_ant = wait_tunnel[-1]
                if last_ant == counter:
                    counter += 1
                    wait_tunnel = wait_tunnel[:-1]
                    nest.append(last_ant)
                else:
                    break
        else:
            wait_tunnel.append(ant_i)
    if len(arrival_queue) == len(nest):
        return 1
    else:
        return 0
```
