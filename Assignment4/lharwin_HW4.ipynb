```python
class Node:
    def __init__(self , data , left=None , right=None):
        self.data = data
        self.left = left
        self.right = right
    
    
    #Function to print the Binary Search Tree in pre-order
    #The function is recursively called until root becomes null
    #The function returs the preorder list
    def preorder_Node(root, list1):
        if not root:
            return 
        list1.append(root.data);
        Left_Node = root.left
        Right_Node = root.right
        if  Left_Node:
            Left_Node.preorder_Node(list1);
        if Right_Node:
            Right_Node.preorder_Node(list1);
        return list1;
    
    
    #Function to print the Binary Search Tree in in-order
    #The function is recursively called until root becomes null
    #The function returs the in-order list
    def inorder_Node(root, list1):
        if not root:
            return 
        Left_Node = root.left
        Right_Node = root.right
        if  Left_Node:
            Left_Node.inorder_Node(list1);
        list1.append(root.data);
        if Right_Node:
            Right_Node.inorder_Node(list1);
        return list1;
    
    
    #Function to print the Binary Search Tree in post-order
    #The function is recursively called until root becomes null
    #The function returs the post-order list
    def postorder_Node(root, list1):
        if not root:
            return 
        Left_Node = root.left
        Right_Node = root.right
        if  Left_Node:
            Left_Node.postorder_Node(list1);
        if Right_Node:
            Right_Node.postorder_Node(list1);
        list1.append(root.data);
        return list1;
    
    
    #Function to print the Binary Search Tree in level order
    #Here we are using the concept of Queue to implement the same. 
    #The function returs the in-order list
    def level_order_Node(root, list1):
        # if root is None, then return nothing
        if root==None:
            return
        #Queue to add the level order nodes
        Queue= []
        # First appending the root node into Queue
        Queue.append(root)
        # the loop continues until all nodes are appended into Queue and later popped out 
        while( len(Queue)!=0):
            #popping the first element from Queue
            first_node = Queue[0];
            #Updating queue after popping
            Queue=Queue[1:]
            if first_node==None:
                continue
            list1.append(first_node.data);
            Queue.append(first_node.left)
            Queue.append(first_node.right)
        return list1
    
    
    #Function to find the maximum node in Binary Search Treee
    def max_Node(root):
        # If root= None, then return nothing.
        # because there is no max Node
        if (root == None):
            return
        # The highest element in BST will always be the element at the rightmost part of the tree
        # So here we are iterating till the rightmost node to find the max Node
        # If there is no Right Node for the Tree, then it return the data in Head node as Head Node is greater than Left Node
        max_Node= root;
        # Iterating till there is no right node
        while(max_Node.right):
            max_Node=max_Node.right
        return max_Node.data;
    
    
    #Function to find the height of the Binary Search Tree
    #Here we recursily call the function on right and left Nodes until root becomes null
    #Finally returs the max of right or left Node. 
    def height_Node(root):
        if root==None:
            return 1
        Left_Node = root.left
        Right_Node = root.right
        # First we check the height of left Node
        if Left_Node:
            Height_Left_Node = Left_Node.height_Node()
        # if left node is null, we randonly assign the height as -1
        else :
            Height_Left_Node =-1
        # Secondly, we check the height of right Node
        if Right_Node:
            Height_Right_Node = Right_Node.height_Node()
        # if right node is null, we randomyl assign the height as -1 
        else:
            Height_Right_Node =-1
        # Finally return the max of right or left Node
        return max(Height_Left_Node, Height_Right_Node) + 1
    
    
    #Function to find if a Node is present in the Binary Search Tree
    #Here we recursively call the function on right and left Nodes until root becomes null
    #Finally returs True or False
    def contains_Node(root,data):
        # if root = None, element is not present, Hence returns false
        if (root == None):
            return False
        Left_Node = root.left
        Right_Node = root.right
        # if element found at root Node, then return True
        if (root.data == data):
            return True
        # if Left_Node is not None, then we recursively call the function until it becomes None
        if Left_Node:
            Left_Node_Element = Left_Node.contains_Node(data)
        # else element is not found in Left_Node
        else:
            Left_Node_Element = False
        # if Right_Node is not None, then we recursively call the function until it becomes None
        if Right_Node:
            Right_Node_Element = Right_Node.contains_Node(data)
        # else element is not found in Right_Node
        else:
            Right_Node_Element= False
        return Left_Node_Element or Right_Node_Element
    
    
    #Function to find the diameter of the Binary Search Tree
    #There are two cases : 
    # Case 1: Longest diameter contains root Node
    # Case 2: Longest diameter is either in left node or right node
    def diameter_Node(root):
        # if root is None return 0
        if root is None:
            return 0
        # if there is no left and right node, then return 0 as there is no diameter
        if root.left == None and root.right==None:
            return 0
        # Case 1 :
        # Assuming that root node is a node in the longest diamater
        # Here we are seperating finding the height of left and right nodes and summing up at the end
        Left_Node= root.left
        Right_Node = root.right
        # Finding height of left node
        if Left_Node:
            Left_Height = Left_Node.height_Node()
        # if left node is None , then height of left node = 0 
        else:
            Left_Height=0
        # Finding height of right node
        if Right_Node:
            Right_Height = Right_Node.height_Node()
        # if right node is None, then height of right node = 0  
        else:
            Right_Height=0
        # Case 2:
        # Assuming that root node is not a node in the longest diameter
        # Here we all diameter_Node() function for left and right node to find if there are any longer diameter 
        # Checking for longer diameter in left node
        # Sub_Left is the length of longest diameter in left sub tree
        if Left_Node:
            Sub_Left = Left_Node.diameter_Node()
        # if left node is null, then Sub_Left is 0
        else:
            Sub_Left=0
        # Checking for longer diameter in left node
        # Sub_Right is the length of longest diameter in right sub tree
        if Right_Node:
            Sub_Right = Right_Node.diameter_Node()
        # if right node is null, then Sub_Right is 0
        else:
            Sub_Right=0
        if Left_Node and  Right_Node:
            add_up= 2
        elif Left_Node or  Right_Node:
            add_up = 1
        else:
            add_up =0
        return max(Left_Height + Right_Height + add_up, Sub_Left , Sub_Right)
            
        
    # Fuction to insert element into binary search tree
    def insert_Node(root, data):
        # if root is None, then we insert new node to root
        if root == None:
            root= Node(data)
            return
        # Creating a node for data
        Insert_Node = Node(data)
        Left_Node = root.left
        Right_Node = root.right
        position = root
        # iterate till we find the position to fit data or till root is None
        while (root != None):
            position = root
            if (data < root.data):
                root = root.left
            else:
                root = root.right
        #position found !!!
        #Checking where to insert
        if (position == None):
            position = newnode
        elif (data < position.data):
            position.left = Insert_Node
        else:
            position.right = Insert_Node
         
        
    # Function to delete an element into binary search tree
    def deleteNode(root, data):
        # if root is None, then we delete nothing
        if root is None:
            return root
        # if node to be deleted is in left sub tree, call delete_Node function for root.left as the element to be deleted is in left sub tree
        if data < root.data:
            root.left = root.left.deleteNode(data)
        # if node to be deleted is in right sub tree, call delete_Node function for root.right as the element to be deleted is in right sub tree
        elif(data > root.data):
            root.right = root.right.deleteNode(data)
        # if node to be deleted is the root node
        else:
            Left_Node = root.left
            Right_Node = root.right
            if Left_Node is None:
                position = Right_Node
                root = None
                return position
            elif Right_Node is None:
                position = Left_Node
                root = None
                return position
            # Finding the node with minimum value
            minimum = root.right
            while(minimum.left is not None):
                minimum = minimum.left
            curr= minimum
            root.data = curr.data
            root.right = root.right.deleteNode(curr.data)
        return root

class BST:
    def __init__(self , node):
        self.root = node
    def insert(self , data):
        if self.root:
            self.root.insert_Node(data)
        else:
            self.root= Node(data)            
    def delete(self , data): 
        self.root.deleteNode(data)
    def preorder(self):
        list1=[]
        self.root.preorder_Node(list1);
        return list1
    def inorder(self):
        list1=[]
        self.root.inorder_Node(list1);
        return list1
    def postorder(self):
        list1=[]
        self.root.postorder_Node(list1);
        return list1
    def level_order(self):
        list1=[]
        self.root.level_order_Node(list1);
        return list1
    def max(self):
        return self.root.max_Node();
    def height(self):
        return self.root.height_Node();
    def __contains__(self , data):
        return self.root.contains_Node(data);
    def diameter (self):
        return self.root.diameter_Node() ;
```
