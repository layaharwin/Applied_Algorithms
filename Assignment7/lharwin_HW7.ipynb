```python
# Question 1:
def merge(arr, l, m, r):
    n1 = m - l + 1
    n2 = r - m
 
    # create temp arrays
    L = [0] * (n1)
    R = [0] * (n2)
 
    # Copy data to temp arrays L[] and R[]
    for i in range(0, n1):
        L[i] = arr[l + i]
 
    for j in range(0, n2):
        R[j] = arr[m + 1 + j]
 
    # Merge the temp arrays back into arr[l..r]
    i = 0     # Initial index of first subarray
    j = 0     # Initial index of second subarray
    k = l     # Initial index of merged subarray
 
    while i < n1 and j < n2:
        if L[i] <= R[j]:
            arr[k] = L[i]
            i += 1
        else:
            arr[k] = R[j]
            j += 1
        k += 1
 
    # Copy the remaining elements of L[], if there
    # are any
    while i < n1:
        arr[k] = L[i]
        i += 1
        k += 1
 
    # Copy the remaining elements of R[], if there
    # are any
    while j < n2:
        arr[k] = R[j]
        j += 1
        k += 1
 # l is for left index and r is right index of the
# sub-array of arr to be sorted
 
def mergeSort(arr, l, r):
    if l < r:
 
        # Same as (l+r)//2, but avoids overflow for
        # large l and h
        m = l+(r-l)//2
 
        # Sort first and second halves
        mergeSort(arr, l, m)
        mergeSort(arr, m+1, r)
        merge(arr, l, m, r)
 
 
def additional_seats(k,h):
    list1 = []
    # Add start and end into list
    for i, j in h:
        list1.append((i, True))
        list1.append((j, False)) 
    list1.sort() 
    output=0
    seats = 0
    
    # Incrementing and decrementing seats according to status
    for _,status in list1:
        if status== True:
            seats+=1
        else:
            seats-=1
        if seats > k:
            if output < seats-k:
                output= seats-k
    if output>0:
        return output
    else:
        return 0
    
print(additional_seats(2, [[0,2],[1,2],[0,3],[2,3]]))
print(additional_seats(3, [[8,10],[2,4],[7,11]]))
print(additional_seats(2, [[1,3],[1,2],[0,4],[0,5]]))
print(additional_seats(2, [[0,4],[0,10],[0,2],[1,2],[2,4],[3,5]]))
```

```python
# Question 2:
import heapq
def minimum_range(lists):    
    max_limit = 0
    # Finding max elemetn from first elements in all lists
    for i in range(0, len(lists)):
        max_limit = max(max_limit , lists[i][0]) 
    out2= max_limit
    # Adding all the first elements into heap
    heap = []
    for i in range(0, len(lists)):
        heapq.heappush(heap, (lists[i][0] , i, 0))
    out1=heap[0][0]
    
    # Iterating till one of the list reaches end element and keeps updating out0 and out1 whenever diff is minimum
    while 1:      
            num,row,col = heapq.heappop(heap)
            length= len(lists[row])-1
            if col == length:
                break     
            max_limit = max( max_limit , lists[row][col+1])
            new_num= (lists[row][col+1], row , col+1)
            heapq.heappush(heap,new_num)
            new_diff= max_limit-heap[0][0]
            out_diff= out2-out1
            if new_diff < out_diff:
                out1=heap[0][0]
                out2=max_limit
    return (out1, out2)

print(minimum_range([[ 3, 6, 8, 10, 15],[ 1, 5, 12 ],[ 4, 8, 15, 16 ],[ 2, 6 ]])) #4,6
print(minimum_range([[ 2, 3, 4, 8, 10, 15 ],[ 1, 5, 12 ],[ 7, 8, 15, 16 ],[ 3, 6 ]])) #4,7
```

```python
# Question 4:
def max_moves(nums):
    nums.sort()
    # Case 1 : when sum of smallest two numbers is less than the greatest number, then we cannot reduce more than the sum of the lowest two numbers
    # Hence we return nums[0]+nums[1]
    if nums[0]+nums[1] < nums[2]:
        return nums[0]+nums[1]
    # Case 2: 
    # Here, lets assume that out is our result
    # And sum is the sum of all the elments in nums i.e sum= nums[0]+ nums[1]+nums[2]
    # Given that we can reduce two elements at a time
    # So, we can mathemarically prove that, sum - 2* out = 1 or 0 
    # therefore, out = ceil([sum/2])
    # Hence we return int((nums[0]+nums[1]+nums[2])/2)
    return int((nums[0]+nums[1]+nums[2])/2)

print(max_moves([2,4,6])) #6
print(max_moves([4,4,6])) #7
print(max_moves([10,1,3]))#4
print(max_moves([4,4,4])) #6
print(max_moves([5,5,5])) #7
print(max_moves([6,6,6])) #9
print(max_moves([1,0,0])) #0
```
