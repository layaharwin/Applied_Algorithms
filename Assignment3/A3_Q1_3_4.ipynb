```python
def commands_count (List):
    # Dictionary to keep track of ls , mv and cp commands
    dict_out={"ls":0, "mv":0, "cp":0}
    for i in List:
        # Incrementing count of commands when they are normally excuted. i.e without !
        if i in dict_out:
            dict_out[i]+=1
        # Checking for commands when they are executed with !
        elif i[0] == '!':
            command = i[1:]
            if int(command)>0:
                if List[int(command)-1]== "ls":
                    dict_out["ls"]+=1
                elif List[int(command)-1]== "mv":
                    dict_out["mv"]+=1
                elif List[int(command)-1]== "cp":
                    dict_out["cp"]+=1
                elif int(command)-1 > 0:
                    # Looping until you get the command 
                    while(List[int(command)-1][0] == '!'):
                        command= List[int(command)-1][1:]
                    if List[int(command)-1]== "ls":
                        dict_out["ls"]+=1
                    elif List[int(command)-1]== "mv":
                        dict_out["mv"]+=1
                    elif List[int(command)-1]== "cp":
                        dict_out["cp"]+=1            
    return list[dict_out["cp"], dict_out["ls"], dict_out["mv"]]
```

```python
print(commands_count([ "ls", "cp", "mv", "mv", "mv", "!1", "!3", "!6"])) 
print(commands_count(["!3", "!4", "cp", "ls", "mv"]))
print(commands_count(["!0", "!4", "cp", "ls", "mv"]))
print(commands_count(["!-1", "!4", "cp", "ls", "mv"]))
print(commands_count(["ls","mv","mv","!6","!2","!8","mv","!1","!2","cp"]))
```

```python
import random
import math

class Node():
    def __init__(its, key, level):
        its.key = key
        # Hold references to node of different level
        its.next = [None]*(level+1)
        
class LookUpSkipList():
    def __init__(its, words, P):
        its.words= words
        # Setting maxlevel as 5 for this example for better visualization of the levels
        its.maxlevel= 5
        its.P= P
        # Creating a dummy node 
        its.head = its.newNode(its.maxlevel, -1)
        # Inially the level is set to 0 
        its.level = 0
        # if more than one word is passed while creating an object, then we insert them one by one into the SkipList
        for key in its.words:
            its.insert(key)  
    
    def insert(its, key):
        skip_list_temp = its.head
        # initializing track array
        track = [None]*(its.maxlevel+1)
        # move to the current reference from top of the skiplist
        for i in range(its.level, -1, -1):
            while skip_list_temp.next[i]!= None and skip_list_temp.next[i].key < key:
                skip_list_temp = skip_list_temp.next[i]
            track[i] = skip_list_temp
        skip_list_temp = skip_list_temp.next[0]
        # Checking if key is found or not
        if skip_list_temp == None or skip_list_temp.key != key:
            # Generate a random level for node
            newlevel = its.generating_level()
            if newlevel > its.level:
                for i in range(its.level+1, newlevel+1):
                    track[i] = its.head
                its.level = newlevel
            # new node created
            n = its.newNode(newlevel, key)
            for i in range(newlevel+1):
                n.next[i] = track[i].next[i]
                track[i].next[i] = n   
    
    def newNode(its, level, value):
        #Function to create a new node for value and level passed to the fucntion 
        curr = Node(value, level)
        return curr
    
    def generating_level(its):
        level = 0
        while random.random()<its.P and \
              level<its.maxlevel:level += 1
        return level
        
    def print(its):
        # Function to print the skip list with all its levels
        print("\n==========Skip List===========")
        skip_list_temp = its.head
        for i in range(its.level+1):
            print("Level",i,":", end=" ")
            level = skip_list_temp.next[i]
            # print words in each level
            if (level!=None):
                while(level != None):
                    #iterating to each work in particular level
                    print(level.key, end=" . ")
                    level = level.next[i]
                print("")

    def delete(its, key):
        skip_list_temp = its.head
        track = [None]*(its.maxlevel+1)
        # move to the current reference from top of the skiplist
        for i in range(its.level, -1, -1):
            while(skip_list_temp.next[i]!= None and skip_list_temp.next[i].key < key):
                skip_list_temp = skip_list_temp.next[i]
            track[i] = skip_list_temp
        skip_list_temp = skip_list_temp.next[0]
        # Checking if node to be deleted is found or not
        if skip_list_temp != None and skip_list_temp.key == key:
            for i in range(its.level+1):
                if track[i].next[i] != skip_list_temp:
                    break
                track[i].next[i] = skip_list_temp.next[i]
            while(its.level>0 and its.head.next[its.level] == None):
                its.level -= 1
    
    # Function to reach current reference from top level.
    def move_till_current_reference(its, skip_list_temp,key):
        for i in range(its.level, -1, -1):
            while(skip_list_temp.next[i]!= None and skip_list_temp.next[i].key < key):
                skip_list_temp = skip_list_temp.next[i]
        return skip_list_temp
    
    def lookup_search(its, key):
        skip_list_temp = its.head
        skip_list_temp = its.move_till_current_reference(skip_list_temp,key)
        skip_list_temp = skip_list_temp.next[0]
        # Checking if key is present or not 
        if skip_list_temp and skip_list_temp.key == key:
            return True
        else:
            its.insert(key)
            return False


lu_sk_lists = LookUpSkipList(["iub","usa","there","sort","god"],0.6)
lu_sk_lists.print()
print("\nTesting insert function: ")
print("Inserting 'word':")
lu_sk_lists.insert("word")
lu_sk_lists.print()
print("\nTesting delete function: ")
print("Deleting 'there':")
lu_sk_lists.delete("there")
lu_sk_lists.print()
print("\nTesting lookup_search function: ")
print("Checking 'iub' in SkipList: ",lu_sk_lists.lookup_search("iub"))
print("Checking 'present' in SkipList: ", lu_sk_lists.lookup_search("present"))
print("Added 'present' to SkipList")
lu_sk_lists.print()

```

```python
def Convert_to_List(input_string):
    str_to_list=[]
    str_to_list[:0]=input_string
    return str_to_list

def reverse_list(input_string):
    return input_string[::-1]

def predict_winner(s):
    # A_stack for tracking apples
    A_stack=[]
    # O_stack for tracking Oranges 
    O_stack =[]
    # Converting str to List 
    s=Convert_to_List(s)
    for i in range(0, len(s)):
        if s[i] == 'A':
            A_stack.append(i)
        elif s[i] == 'O':
            O_stack.append(i)
    A_stack= reverse_list(A_stack)
    O_stack= reverse_list(O_stack)
    i=0
    for i in range(0, len(s)):
        if s[i] == -1:
            continue
        if s[i] == 'A':
            if O_stack:
                j = O_stack.pop()
                s[j] =-1
        elif s[i] == 'O':
            if A_stack:
                j = A_stack.pop()
                s[j] = -1
    if A_stack:
        return 'apple'
    else:
        return 'orange'

```

```python
print(predict_winner("AOOAOAAOOAAOAOOA"))
print(predict_winner("AOO"))
print(predict_winner("AO"))
print(predict_winner("AAAAAOOOOOOO"))
print(predict_winner("A"))
```
